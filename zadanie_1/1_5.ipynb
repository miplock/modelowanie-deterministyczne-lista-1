## 1.5.

>**Polecenie:** Zaobserwuj na wykresie punkty stałe, punkty podwojenia okresu (bifurkacje) oraz obszary, w których zachowanie wydaje się chaotyczne.

**Rozwiązanie:**  

Obserwujemy następujące etapy zmian dynamiki układu:

- Dla $0 < r < 1$: stabilny punkt $x^* = 0$.  
- Dla $1 < r < 3$: stabilny punkt $x^* = 1 - \frac{1}{r}$, widoczny jako pojedyncza gałąź rosnąca do ok. $x=0.66$ przy $r=3$.

W punkcie $r = 3$ pojawia się pierwsze **podwojenie okresu** (okres 2).  
Kolejne występują przy $r \approx 3.45$, $3.54$, $3.56$, prowadząc do chaosu przy
$$
r_\infty \approx 3.6.
$$

Dla $r > r_\infty$ układ staje się chaotyczny, a wartości $x_n$ tworzą gęstą chmurę.  

Wraz ze wzrostem $r$ układ przechodzi od:
- stanu stacjonarnego ($r<3$),  
- przez kolejne podwojenia okresu,  
- do chaosu z lokalnymi wyspami regularności.
