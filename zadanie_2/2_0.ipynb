## 2.0.

>**Polecenie:** Opisz co oznaczają parametry modelu. Znajdź stany stacjonarne analitycznie. Przy jakich założeniach na parametry istnieją dwa stany stacjonarne?

**Rozwiązanie:**

**Wyjaśnienie parametrów modelu:**
- $N(t)$ — liczebność populacji w czasie $t$,
- $r>0$ — tempo wzrostu wewnętrznego,
- $K>0$ — pojemność środowiska,
- $k\geq 0$ — intensywność proporcjonalnego odłowu.

**Stany stacjonarne:**

$$f(N) = 0$$
$$rN\left(1-\frac NK\right) -kN = 0$$
$$
N_1^* = 0, \qquad N_2^* = K\left(1 - \frac kr\right).
$$
Zakładamy, że populacja jest rzecz jasna nieujemna, więc drugi stan stacjonarny istnieje, gdy $k < r$.
