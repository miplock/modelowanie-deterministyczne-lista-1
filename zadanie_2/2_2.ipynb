## 2.2.

>**Polecenie:** Użyj parametru $K = 1$ i zbadaj relację "$r$ vs $k$": (a) $r = 1$, $k = 0.5$ oraz (b) $r = 1$, $k = 1.5$. Jak się zmieniają stany stacjonarne? Możesz posłużyć się wyprowadzeniem analitycznym.

**Rozwiązanie:**

### 2.2.a.

$$
\frac{dN}{dt} = 1 \cdot N\left(1-\frac N1\right) -0.5\cdot N = N(0.5 - N)
$$
Mamy rzecz jasna $N_1^*=0$. Drugim stanem stacjonarnym jest $N_2^*=0.5$. Jako że $k<r$ i stąd $N_2^*\geq 0$ to jest sens mówić o tym rozwiązaniu. 

### 2.2.b.

$$
\frac{dN}{dt} = 1 \cdot N\left(1-\frac N1\right) -1.5\cdot N=N(-0.5-N)
$$

Ponownie mamy $N_1^*$. Prawa strona równania jest równa zeru również dla $N\equiv-0.5$, natomiast nie ma to sensu, by mówić o ujemnej populacji. Zgodnie z [2.0.](./2_0.ipynb) widzimy więc, że mamy jedno rozwiązanie stacjonarne w sytuacji, gdy $k < r$.
