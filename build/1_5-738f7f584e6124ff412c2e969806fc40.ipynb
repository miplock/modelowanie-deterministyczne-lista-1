## 1.5.

>**Polecenie:** Zaobserwuj na wykresie punkty stałe, punkty podwojenia okresu (bifurkacje) oraz obszary, w których zachowanie wydaje się chaotyczne.

**Rozwiązanie:**  

### 1. Punkty stałe

- \(x^\* = 0\) — **niestabilny** dla każdego \(r > 0\), ponieważ  
  \(|f'(0)| = |1 + r| > 1\).
- \(x^\* = 1\) — **stabilny** dla \(0 < r < 2\).  
  W punkcie \(r = 2\) spełniony jest warunek \(|f'(x^\*)| = 1\), co oznacza **utracenie stabilności** i początek oscylacji okresu 2.

---

### 2. Podwojenia okresu (bifurkacje Feigenbauma)

- pierwsze: \(r_1 = 2.000\) — przejście ze stanu stałego do cyklu o okresie 2,  
- kolejne (odpowiedniki klasycznych wartości dla logistycznego \(\mu\) pomniejszone o 1):  
  \(r_2 \approx 2.44949\),  
  \(r_3 \approx 2.54409\),  
  \(r_4 \approx 2.56441\),  
  …  
- kumulacja przy \(r_\infty \approx 2.56995\).  
  Wraz z każdym kolejnym progiem liczba gałęzi (okres) podwaja się: 2, 4, 8, 16, …

---

### 3. Obszar chaosu

- Dla \(r > r_\infty \approx 2.56995\) układ przechodzi w **chaos deterministyczny**.  
  Wykres pokazuje wtedy gęsty, rozproszony zbiór punktów.  
- W chaosie występują **okna okresowe**:
  - **okno okresu 3** około \(r \approx 2.82843\),  
  - mniejsze okna innych okresów (5, 6, 7, …) w zakresie \(r \in (2.57, 3]\).

---

### 4. Dodatkowe obserwacje

- **Scalanie i rozszczepianie pasm** w chaosie (ang. *band merging/splitting*).  
- Dla \(r \to 3\) wartości \(x_n\) wypełniają coraz większą część przedziału \((0, 1)\).

---

**Podsumowanie:**  
Na wykresie widoczne są:
1. dwa punkty stałe (0 — niestabilny, 1 — stabilny do \(r=2\));  
2. kaskada podwojeń okresu do \(r_\infty \approx 2.57\);  
3. obszary chaosu z oknami okresowymi.
